```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv('Datasel_lab3.csv')
df
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df
```

```python
# Strategy 1: Mean Imputation (for 'Size_SqFt')
imputer_mean = SimpleImputer(strategy='mean')
df['Size_SqFt'] = imputer_mean.fit_transform(df[['Size_SqFt']])
print("'Size_SqFt' imputed with MEAN.")
```

```python
df
```

```python
# Strategy 2: Median Imputation (for 'Num_Bedrooms')
imputer_median = SimpleImputer(strategy='median')
df['Num_Bedrooms'] = imputer_median.fit_transform(df[['Num_Bedrooms']])
print("'Num_Bedrooms' imputed with MEDIAN.")
df
```

```python
# Strategy 3: Mode Imputation (for 'Has_Garage')
imputer_mode = SimpleImputer(strategy='most_frequent')
df['Has_Garage'] = imputer_mode.fit_transform(df[['Has_Garage']]).ravel()
print("'Has_Garage' imputed with MODE (most_frequent).")
```

```python
df
```

```python
df.info()
```

```python
# Strategy 1: Label Encoding (for binary 'Yes/No' columns)
# Converts 'Yes'/'No' to 1/0. Good for binary or ordinal data.
le = LabelEncoder()

# Encode 'Has_Garage'
df['Has_Garage'] = le.fit_transform(df['Has_Garage'])
df
```

```python
# Encode 'Is_Newly_Built'
df['Is_Newly_Built'] = le.fit_transform(df['Is_Newly_Built'])
df
```

```python
# 1. Initialize the encoder
# We add sparse_output=False to get a regular array
# We add drop='first' to avoid the dummy variable trap (good practice)
encoder = OneHotEncoder(sparse_output=False, drop='first')

# 2. Fit and transform ONLY the 'Location' column
data_encoded_array = encoder.fit_transform(df[['Location']])

# 3. Get the new column names
column_names = encoder.get_feature_names_out(['Location'])

# 4. Create a new DataFrame with the encoded data
# We pass df.index to make sure the concatenation in step 6 works
data_encoded_df = pd.DataFrame(data_encoded_array, columns=column_names, index=df.index)

# 5. Drop the original 'Location' column from the main DataFrame
data_remaining = df.drop(columns=['Location'])

# 6. Concatenate the remaining data with the new encoded DataFrame
df = pd.concat([data_remaining, data_encoded_df], axis=1)
```

```python
df
```

```python
X = df.drop(['Is_Newly_Built', 'House_Price'], axis=1)
y = df['Is_Newly_Built']
```

```python
X
```

```python
y
```

```python
# 1. Visualize Data BEFORE SMOTE
print("Visualizing imbalance BEFORE SMOTE...")
plt.figure(figsize=(7, 5))
sns.countplot(x=y)
plt.title('Count of Target (Is_Newly_Built) - BEFORE SMOTE')
plt.xlabel('Class (0 = No, 1 = Yes)')
plt.ylabel('Frequency')
plt.savefig('smote_before.png')
print("Saved 'smote_before.png'")
plt.show()

print("\nValue counts BEFORE SMOTE:")
print(y.value_counts())
```

```python
# 2. Apply SMOTE
# SMOTE creates new synthetic samples of the minority class (1).
smote = SMOTE(random_state=42, k_neighbors=2)

X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# 3. Visualize Data AFTER SMOTE
print("Visualizing balance AFTER SMOTE...")
plt.figure(figsize=(7, 5))
sns.countplot(x=y_resampled)
plt.title('Count of Target (Is_Newly_Built) - AFTER SMOTE')
plt.xlabel('Class (0 = No, 1 = Yes)')
plt.ylabel('Frequency')
plt.savefig('smote_after.png')
print("Saved 'smote_after.png'")
plt.show()
```

```python
print("\nValue counts AFTER SMOTE:")
print(pd.Series(y_resampled).value_counts())
```

```python
# We will separate our regression target 'y' (House_Price) first.
# We also drop 'Is_Newly_Built' as it was our classification target.
feature_columns = df.drop(['House_Price', 'Is_Newly_Built'], axis=1).columns
X_features = df.drop(['House_Price', 'Is_Newly_Built'], axis=1)

print(f"Scaling the following {len(feature_columns)} features:")
print(list(feature_columns))

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the features
X_scaled = scaler.fit_transform(X_features)

# Convert the scaled numpy array back to a DataFrame for easy viewing
df_scaled_features = pd.DataFrame(X_scaled, columns=feature_columns)

print("\nScaled Features (Head):")
df_scaled_features.head()
```

```python
print("--- Part 7: Distance Metrics Demonstration ---")

# Let's use the first two data points (rows) from our scaled data.
# We use .values to get them as numpy arrays.
p1 = df_scaled_features.iloc[0].values
p2 = df_scaled_features.iloc[1].values

def euclidean_distance(v1, v2):
    """Calculates Euclidean distance (L2 norm) from scratch."""
    return np.sqrt(np.sum((v1 - v2)**2))

def manhattan_distance(v1, v2):
    """Calculates Manhattan distance (L1 norm) from scratch."""
    return np.sum(np.abs(v1 - v2))

# Calculate the distances
euclidean = euclidean_distance(p1, p2)
manhattan = manhattan_distance(p1, p2)

print(f"Sample Point 1 (scaled): {np.round(p1, 2)}")
print(f"Sample Point 2 (scaled): {np.round(p2, 2)}")
print(f"\nEuclidean Distance (L2): {euclidean:.4f}")
print(f"Manhattan Distance (L1): {manhattan:.4f}")
```
